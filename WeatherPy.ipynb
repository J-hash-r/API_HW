## Linear Regression

```python
Southern = WeatherDF.loc[WeatherDF["Latitude"] < 0]
Northern = WeatherDF.loc[WeatherDF["Latitude"] > 0]
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python

ax = Northern.plot(kind='scatter', x="Latitude", y="Temperature [°C]")

# Set the x-axis label
ax.set_xlabel("Northern Latitude")
x=Northern["Latitude"]
y=Northern["Temperature [°C]"]
# Set the y-axis label
ax.set_ylabel("Temperature")
ax.set_title("Northern Latitude vs Temperature")

#regression
from scipy import stats
(slope, intercept, r_value, p_value, std_err) = stats.linregress(x, y)
Northern_LinReg1 = intercept + slope * x
plt.plot(x,Northern_LinReg1,"-r")
Northern_LinRegEQ = "Y = " + str(round(slope,2)) + "X +"+ str(round(intercept,2)) + "  Pearson Corellation Coefficient = " + str(round(r_value,2))

plt.annotate(Northern_LinRegEQ,(70,0),fontsize=10,color="black")

plt.savefig('Northern Latitude vs Temperature.png')
```
